```python
class rotationSystem:

    def __init__(self, vertex, edge):
        """Takes in two lists of tuples of half edges"""
        self.vertex = vertex
        self.edge = edge

    def findHalfEdgeRotation(self, list1, x):
        """Find half edge rotation"""
        for l in list1:
            if x in l:
                for i in range(len(l)):
                    if l[i] == x:
                        halfEdge = l[(i+1) % len(l)]
        return halfEdge

    def findHalfEdgeEdge(self, list1, x):
        """Find edge permutation"""
        for l in list1:
            if x in l:
                if l[0] == x:
                    return l[1]
                else:
                    return l[0]

    #find walk - for one edge
    def findWalk(self, permutation, edges, o):
        """Find border walk"""
        #set original half edge
        startEdge = o
        cycleList = [o]
        currentEdge = self.findHalfEdgeRotation(permutation, o)
        cycleList.append(currentEdge)
        currentEdge = self.findHalfEdgeEdge(edges,currentEdge)
        cycleList.append(currentEdge)
        # while not at original half edge:
        #do an and to cut out above 4 lines
        while currentEdge != o:
            #get half edge rotation
            #add to cycle list
            # get half edge edge
            # add to cycle list
            currentEdge = self.findHalfEdgeRotation(permutation, currentEdge)
            cycleList.append(currentEdge)
            currentEdge = self.findHalfEdgeEdge(edges,currentEdge)
            cycleList.append(currentEdge)
        #return cycle list
        return cycleList


    def findAllWalks(self):
        """Find all border walks"""
        checked = []
        lenCycles = []
        for i in range(len(self.edge)):
            for j in range(2):
                if self.edge[i][j] not in checked:
                    walk = self.findWalk(self.vertex, self.edge, self.edge[i][j])
                    
                    lenCycles.append(walk)
                    for x in range(0, len(walk),2):
                        checked.append(walk[x])
                    # print(checked)
        for l in lenCycles:
            for k in lenCycles:
                if k != l:
                    if set(k) == set(l):
                        lenCycles.remove(k)
                        
        return lenCycles

    def displayCycles(self):
        for i in range(len(self.noCycles())): #print out number of each cycle
            if self.noCycles()[i] != 0: #prevents printing out number of k-cycles if there is no k-cycle 
                print(f"Number of {i}-cycles is: {self.noCycles()[i]}")

```
